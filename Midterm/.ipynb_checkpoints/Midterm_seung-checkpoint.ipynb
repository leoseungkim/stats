**2. Evaluate the assumptions of normality and homogeneity of variance, including any relevant plots. What do you conclude about these assumptions?**

**Solution.** $F_{observed}$ follows an $F$ distribution only if the assumptions of normality and homogeneity of variance are met, in addition to the assumption that scores are independent of one another. Briefly, 

* **Assumption of normality** posits that the distribution of scores within each group must be normal;
* **Assumption of homogeneity of variance** posits that the population variance of each group must be equal to each other's.

Let's take a look at normality using the Shapiro-Wilk test, Q-Q plots, and histograms:

```r
nor.test(Distress ~ Condition, data=lepore, method='SW', plot='qqplot-histogram')
```

OK, according to the Shapiro-Wilk test, each group can be considered normal. Even just visually, we see that none of the points on the Q-Q Plots radically depart from the line (which represents a perfectly normal distribution), so that serves as pretty good evidence for normality. Some of the histograms look a little wonky (especially NoTalk and Talk), but they look better with a little fussing with the bin width:

```r
ggplot(lepore,aes(x=Distress))+geom_histogram()+facet_grid(~Condition)+stat_bin(binwidth=0.5)+theme_bw()
```

As for the homogeneity of variance, in the case of equal $n$, as long as the ratio of the largest sample group variance to the smallest sample group variance remains below 4, the ANOVA will tolerate moderate violations. But it looks like there isn't *that* much of a violation anyway. We are all good:

$$\frac{s^2_{max}}{s^2_{min}} = \frac{0.72}{0.45}=1.6 < 4$$

**3. Using R, conduct the omnibus test (within a GLM framework) that all group means are the same. Write out the full and restricted models using the cell means parameterization, identifying the meaning of each term. What conclusion can you draw from the omnibus test?**

**Solution.** Let's articulate the models using the cell means parameterization first. 

* Full: $Y_{ij} = \mu_{j}+\epsilon_{ijF}$
* Restricted: $Y_{ij} = \mu+\epsilon_{ijR}$ where $\mu=3.12$

In the models, 

* The $Y_{ij}$ refers to each individual observation, where $i$ indexes the within-group number of the observation (e.g. '*5th observation* in NoTalk group') and $j$ indexes which group the observation belongs to. 
* The $\mu$ is the 'cell mean' for each unique group considered—the full model has $\mu_1$ through $\mu_4$ since it considers four different groups (NoTalk, Talk, Validate, Challenge), while the restricted model only has a single $\mu$ (the grand mean) since it assumes that there is no difference between the four groups.
* The $\epsilon_{ij}$ is the error term. Each individual observation has its own error term—the error term is simply the difference between the observation and the cell mean, or the portion of the observation that cannot be explained by the cell mean.

If I understand correctly from your comments on HW2, if the experiment design is unbalanced, R will take as the grand mean $\mu$ for an unweighted average of the group averages rather than the average of all observations. **Q: is R right, or should we correct R?** In any case, since our design is balanced ($n=20, N=80$), we don't have to worry about this.

```r
mean(lepore$Distress)
```

Instead of calculating everything out by hand, we'll try out fitting models with `aov()`.

```r
full = aov(Distress ~ Condition, data=lepore)
summary(full)
```

```r
restr = aov(Distress ~ 1, data=lepore)
summary(restr)
```

Then we'll plug in the models into a comparison using `anova()`.

```r
anova(restr, full)
```

So we get $F=5.007$ and a corresponding $p=0.003$ which is below $\alpha=0.05$. So, there is enough evidence for us to reject the null hypothesis (that there is no difference between the group means)—the group means are not all the same.

**5. Suppose you are interested in whether there is an effect of talking about one’s thoughts and feelings, whether to oneself or to a confederate, compared to not talking. Come up with the appropriate contrast coefficients and test the contrast for significance using R. Report the point estimate and 95% confidence interval for the mean difference. What do you conclude?**

**Solution.** This is a complex comparison. Essentially, we want to compare the mean of NoTalk to the mean of Talk, Validate, and Challenge combined. Our null hypothesis is that there is no difference between the mean of NoTalk and the mean of Talk, Validate, and Challenge ($\mu_1 = \frac{1}{3}(\mu_2+\mu_3+\mu_4)$). So, the models we can write are:

* Full: $Y_{ij} = \mu_j + \epsilon_{ij}$
* Restricted: $Y_{ij} = \mu_j + \epsilon_{ij}$ where $\mu_1 = \frac{1}{3}(\mu_2+\mu_3+\mu_4)$

So let's get the group means again:

```r
describeBy(lepore$Distress, lepore$Condition)
```

If the null hypothesis is true, we would expect $\psi=0$. The task is to figure out whether $\hat{\psi}$ is statistically different from 0. We can assign coefficients such that $c_1 = \frac{1}{3}(c_2+c_3+c_4)$ and get $\hat{\psi}$ as below:

$$\hat{\psi} = 1(\mu_1) + \Big(-\frac{1}{3}\Big)\mu_2 + \Big(-\frac{1}{3}\Big)\mu_3 + \Big(-\frac{1}{3}\Big)\mu_4$$
$$=3.44 - \frac{3.28}{3} - \frac{3.03}{3} - \frac{2.73}{3}$$
$$=0.427$$

So the mean Distress score for NoTalk is slightly higher (0.427) than the mean Distress score for Talk, Validate, and Challenge combined. The task is to figure out whether this difference is significant. 

Since 

$$SS_{between} = E_R-E_F = \frac{\hat{\psi}^2}{\sum_{j=1}^a\Big(\frac{c_j^2}{n_j}\Big)}$$

```r
c_sq = (-1/3)^2
SS_between = 0.427^2/(1/20 + c_sq/20 + c_sq/20 + c_sq/20)
SS_between
```

Our $df_F = 76$ since we estimate four $\mu$s; $df_R = 77$ since we only estimate three (we assume that $\mu_1 =\frac{\mu_2+\mu_3+\mu_4}{3}$). So $df_R-df_F = 1$.

```r
MS_within = (0.45^2 + 0.72^2 +0.66^2 + 0.59^2)/4
MS_within
```

```r
F = (SS_between/1)/MS_within
F
```

```r
qf(0.05, 1, 76, lower.tail=FALSE)
```

Since $F_{obs}=7.271$ and $F_{0.05;1,76} = 3.967$, our $F_{obs}$ is significant. 

Let's get the confidence interval for $\psi$:

$$\hat{\psi} \pm \sqrt{F_{\alpha;1,N-a}}\sqrt{MS_{within}\sum_{j=1}^a\Big(\frac{c^2_j}{n_j}\Big)}$$

```r
psi = 0.427
F = 3.967
MS_within = 0.37615
sum = (1/20 + c_sq/20 + c_sq/20 + c_sq/20)

psi+(sqrt(F)*sqrt(MS_within*sum))
psi-(sqrt(F)*sqrt(MS_within*sum))
```

So the confidence interval for $\psi$ is $[0.112, 0.742]$. This is to be expected, since the $F$ test showed that the difference is significant. The mean difference is nonzero. 

**6. Test whether there is a difference between the validating and challenging conditions. Report the point estimate and 95% confidence interval for the mean difference. What do you conclude?**

**Solution.** Our null hypothesis is that there is no difference between the means of the Validate and Challenge conditions ($H_0: \mu_3 = \mu_4$). So, the models we can write are:

* Full: $Y_{ij} = \mu_j + \epsilon_{ij}$
* Restricted: $Y_{ij} = \mu_j + \epsilon_{ij}$ where $\mu_3=\mu_4$

```r
describeBy(lepore$Distress, lepore$Condition)
```

The null hypothesis $H_0: \mu_3=\mu_4$ equivalent to

$$(0)\mu_1 + (0)\mu_2 +(1)\mu_3 + (-1)\mu_4 = 0$$

The left-hand expression is what we call the contrast ($\psi$). So:

$$\hat{\psi} = 3.03 - 2.73 = 0.3$$

Cool. So now, we can get our $SS_{between}$:

$$E_R-E_F = \frac{\hat{\psi}^2}{\sum_{j=1}^a\Big(\frac{c^2_j}{n_j}\Big)} = \frac{0.3^2}{\frac{0^2}{20}+\frac{0^2}{20}+\frac{1^2}{20}+\frac{(-1)^2}{20}} = \frac{0.09}{\frac{1}{10}}$$

$$=0.9$$

$df_R = 77$ because we are estimating three $\mu$s ($\mu_1, \mu_2, \mu_3=\mu_4$); $df_F = 76$ since we are estimating four ($\mu_1, \mu_2, \mu_3, \mu_4$). So $df_R-df_F=1$. So $MS_{between} = 0.9/1 = 0.9$.

$$MS_{within} = \frac{E_F}{df_F} = \frac{0.45^2+0.72^2+0.66^2+0.59^2}{4} = \frac{0.2025+0.5184+0.4356+0.3481}{4}$$
$$=\frac{1.5046}{4} = 0.37615$$

So $F = 0.9/0.37615 = 2.393$, which is lower than 3.967. There is no difference between the Validate and Challenge conditions. 

Let's do the confidence interval now:

```r
psi = 0.3
F = 3.967
MS_within = 0.37615
sum = 1/10

psi+(sqrt(F)*sqrt(MS_within*sum))
psi-(sqrt(F)*sqrt(MS_within*sum))
```

So the confidence interval for $\psi$ is $[-0.0863, 0.686]$. This is, again, expected—the difference is not significant, and the interval contains 0.

$$R^2_{effect\ size} = \frac{SS(\psi)}{SS_{total}} = \frac{SS(\psi)}{SS_{between}+SS_{within}} = \frac{2.734935}{5.764+}$$

Since $MS_{within} = 0.37615 = \frac{SS_{within}}{df_{within}}$

$R^2_{alerting}$ is larger than $R^2_{effect\ size}$ since $R^2_{alerting}$ describes the ratio between the specific contrast and the between-group variability, while $R^2_{effect\ size}$ describes the ratio between the specific contrast and the total variability in the data. **A researcher is interested in what therapeutic strategies will best engage refugees in a treatment for post-traumatic stress disorder (PTSD). The researcher recruited 18 refugees and randomly assigned them to one of 3 equally sized groups. In each group, a different semi-structured interview was administered as if it were the first therapy session, and then the client was asked to rate on a scale from 1 to 100 how engaged they felt, where higher scores indicate feelings of greater engagement. The first group was given a standard PTSD semi-structured intake interview, the second group was given an interview that focused on post-traumatic growth, and the third group was given an interview that focused on post-traumatic growth and was culturally adapted. Summary statistics for the data are presented below.**

**11. Using the effects parameterization, state the full and restricted models being compared in an omnibus one-way ANOVA of these data. Report the estimated parameter values (1 $μ$ and 3 $α_j$s), and the degrees of freedom associated with each model.**

**Solution.** OK, so these are equally sized groups. We can simply take the mean of sample means to figure out the grand mean:

$$\mu = \frac{45+53+64}{3} = 54$$

So here are the models:
* Full: $Y_{ij} = 54 + \alpha_j+\epsilon_{ijF}$
* Restricted: $Y_{ij} = 54 + \epsilon_{ijR}$ 

For the Full model, $\alpha_1 = -9, \alpha_2 = -1, \alpha_3 = 10$, and $df_F$ = 18-3 = 15 since we are estimating three values ($\mu$ and two of the $\alpha$s). For the Restricted model, $df_R=18-1=17$ since we only estimate a single grand mean $\mu$. 

OK, let's bang this bad boy out. 

$$E_R-E_F = SS_{between} = 6(-9)^2 + 6(-1)^2+6(10)^2 = 1092$$

$$df_R-df_F = 32-30 =2$$

So,

$$MS_{between} = \frac{E_R-E_F}{df_R-df_F} = \frac{1092}{2} = 546$$

And

$$MS_{within} = \frac{E_F}{df_F} = \frac{\sum s^2_j}{a}= \frac{10^2 + 11^2 + 12^2}{3} = 121.667$$

So,

$$F = \frac{MS_{between}}{MS_{within}} = \frac{546}{121.667} = 4.488$$

```r
qf(0.05, 2, 15, lower.tail=FALSE)
```

Since $F=4.488$ is more extreme than $F_{0.05;2,15}=3.682$, the test is significant. There *is* a difference among the means of the three groups.
